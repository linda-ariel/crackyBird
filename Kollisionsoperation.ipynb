```c++
//Kollisionsoperationen

//mit Protagonist
void kollision(Protagonist p, Item i) {
    int Xabstand = i.getXpos() - p.getX(); //positiv wenn p vor i
    int Yabstand = i.getYpos() - p.getY(); //positiv wenn p über i
    //mit Münze -> bringt 1 Punkt
    if (i.getArt() == 1) {
        if (abs(Xabstand) < 55 && abs(Yabstand) < 55) {//55->Breite/Höhe Item(15) + Breite/Höhe Protagonist(40) / *wenn sich Größe bei ducken ändert -> extra Bedingung mit erfassen von Bildwert* 
            münzeGesammelt(i);
            p.setPunkte(1);
            //Punkte dazu da, mehr Leben zu sammeln -> Level hat feststehende Länge?
            //schon in Operation setPunkte() einbauen?
            if(p.getPunkte() > 20) { // ein Leben = 20pkt , //getPunkte & getLeben fehlen noch in Klasse Protagonist
                 p.setLeben(1)
                 p.setPunkte(-p.getPunkte()); //Punktestand wird wieder auf null gesetzt 
            }
        }
    }
    
    //mit Feind
    if (i.getArt() == 2) {
        //wenn Protagonist gegen Feind läuft -> verliert 1 Leben und 5 Punkte
        if (Xabstand == 70 && abs(Yabstand) < 90) { //Davon ausgehend, dass Feind sich immer nur 1 Pixel nach links/oben/unten bewegt 
            p.setLeben(-1);
            p.setPunkte(-5);
            //Schon in Operation setPunkte() einbauen?
            if (p.getPunkte() < 0) {
                p.setPunkte(-p.getPunkte());
            } 
        //wenn Protagonist Feind besiegt (auf ihn springt) -> bringt 5 Punkte
        } else if (abs(Xabstand) < 70 && Yabstand == 90) {
            feindTot(i);
            p.setPunkte(5);
            //if (Punkte > 20)...  
        }
        
    }
    
    //mit Stein
    if (i.getArt() == 0) {
        //wenn Protagonist gegen Stein läuft -> verliert 3 Punkte
        if (Xabstand == 65 && abs(Yabstand) < 65) {
            p.setPunkte(-5);
        //wenn Protagonist auf von unten gegen Stein springt -> bringt 3 Punkte
        } else if (abs(Xabstand) < 65 && Yabstand == -65) {
            steinZerstört(i);
            p.SetPunkte(3);
            //if (Punkte > 20) ...
        //wenn protagonist auf Stein springt
        } else if (abs(Xabstand) < 65 && Yabstand == 65) {
            p.setX(i.getXpos() - 65); //fehlt noch in Klasse Protagonist
        }
    }
            
}



// zu Klasse Protagonist hinzufügen ?
//1. bei setPunkte überprüfen, ob Protagonist mehr als 20 oder weniger als 0 punkte hat
//2. setX() und setY()
```
