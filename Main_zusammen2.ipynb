```c++
#include <algoviz/SVG.hpp>
#include "Level2.hpp"
#include "Protagonist.hpp"
#include "Hindernis.hpp"

AlgoViz::clear();

SVG view = SVG(700, 500, "Level 1");
Image backround = Image("erste_uni.jpg",350,250,700,500,&view);
view.drawRect(0,0,700,550);
view.drawRect(110,0,80,550);

view.setStatusMsg("Drücken Sie x zu beenden!");

Text anzeigeL = Text("Leben: 3", 15, 20, &view);
Text anzeigeP = Text("Punkte: 0", 15, 37, &view);
view.drawRect(110,380,80,80);


MouseState maus;

//Spieler
Protagonist mario = Protagonist(&view, &anzeigeP, &anzeigeL,1);


//Karte des Levels
Level levelMap = Level(1,5,5,&view);

bool blockade = false; // Einmal gedrückt, ist der Sprung vorerst nicht beeinflussbar...man kann aber "Supersprung" aktivieren
bool blockade2 = false; // Blockiert jegliche Aktionen während des Runterspringens
int arrowUp = 0; // Speichert die Anzahl der "arrowUps", die während der Blockade gedrückt werden...
bool supersprungB = false;
string taste = "";


int i = 0;
int zaehler = 0;
int yOben;
int yObenFix;
double vZaehler = 2;
double zaehlerUnten1 = 0;
double zaehlerUnten2 = 0;
bool dreihundert = false;

while ( taste != "x" ) {
    
   if ( zaehlerUnten1 >= 500 ) {
        
        if (levelMap.getEbene2() == 1) {
            levelMap.add(1);
        }
        zaehlerUnten1 = 0;
        dreihundert = true;
    }
    
    if ( zaehlerUnten1 >= 300 && dreihundert == true ) {
        
        if (levelMap.getEbene2() == 1) {
            levelMap.add(1);
        }
        zaehlerUnten1 = 0;
        dreihundert = false;
    }

    //Wenn die Karte zu ende es, wird sie von vorne wiederbefüllt
    if ( zaehlerUnten2 >= 400 ) {
        
        zaehler++;
        if ( zaehler == 30 ) {
            cout << "Level up!" << endl;
            zaehler = 0;
            levelMap.fuellMap();
            double neuV = levelMap.getV() + 0.25;
            vZaehler = neuV;
            levelMap.setV(neuV);
            mario.setV(neuV);
        }
        if (levelMap.getEbene1() == 1) {
            levelMap.add(0);
        }
        
        /*if (levelMap.getEbene2() == 1) {
            levelMap.add(1);
        }*/
            
        if (levelMap.getEbene3() == 1) {
            levelMap.add(2);
        }
        
        zaehlerUnten2 = 0;
 
    }
    
    
    /*für die drei Unterschiedliche Höhe
    for(int j = 0; j < 3; j++){
        
        
        switch(levelMap.getValue(i,j)){
            case 0 : 
                levelMap.move();
                break;
            case 1:
                levelMap.add(j);  
                levelMap.move();
                break;
                
        }*/
        
       /* if (levelMap.getValue(i,j) == 1) { 
            
            levelMap.add(j);  
            levelMap.move();
            
            
        } else {
            
            levelMap.move();
          
        }
    }*/
    
    levelMap.move();
    
    taste = view.lastKey();
    yOben = levelMap.woIstBoden(mario.getY());
    mario.setYoben(yOben);

    if ( taste == "ArrowUp" && blockade == false ) {
        yObenFix = yOben;
        if ( mario.getBildwert() == 2 ) {
            mario.switchBildaktiv("ArrowUp");
        } else if ( supersprungB == false && mario.getY() + 50 >= yObenFix - 45 ) { 
            mario.springenhoch();
            blockade = true; 
        }
    } else if ( taste == "ArrowUp" && blockade == true ) {
        if ( mario.getY() + 50 >= yObenFix - 45 && supersprungB == false && blockade2 == false ) {
            mario.springenhoch();
            arrowUp++;
            if ( mario.getY() + 50 < yObenFix - 45 && arrowUp >= 1 ) { 
                arrowUp = 0;
                supersprungB = true;
                mario.supersprung();
            } else if ( mario.getY() + 50 < yObenFix - 45 && arrowUp < 1 ) {
                arrowUp = 0;
                blockade2 = true;
            } 
        } else if ( mario.getY() + 280 >= yObenFix - 45 && supersprungB == true && blockade2 == false ) {
            mario.springenhoch();
            if ( mario.getY() + 280 < yObenFix - 45 ) {
                blockade2 = true;
            }
        } else if ( mario.getY() < yOben - 40 && blockade2 == true ) {
            mario.springenrunter();
            if ( mario.getY() >= yOben - 40 ) {
                blockade2 = false;
                blockade = false;
                supersprungB = false;
                mario.setSupersprung(false);
            }
        }
    } else if ( taste != "ArrowUp" && blockade == true ) {
        if ( mario.getY() + 50 >= yObenFix - 45 && supersprungB == false && blockade2 == false ) {
            mario.springenhoch();
            if ( mario.getY() + 50 < yObenFix - 45 && arrowUp >= 1 ) {
                arrowUp = 0;
                supersprungB = true;
                mario.supersprung();
            } else if ( mario.getY() + 50 < yObenFix - 45 && arrowUp < 1 ) {
                arrowUp = 0;
                blockade2 = true;
            }
        } else if ( mario.getY() + 280 >= yObenFix - 45 && supersprungB == true && blockade2 == false ) {
            mario.springenhoch();
            if ( mario.getY() + 280 < yObenFix - 45 ) {
                blockade2 = true;
            }
        } else if ( mario.getY() < yOben - 40 && blockade2 == true ) {
            mario.springenrunter();
            if ( mario.getY() >= yOben - 40 ) {
                blockade2 = false;
                blockade = false;
                supersprungB = false;
                mario.setSupersprung(false);
            }
        }
    } else if ( taste == "ArrowDown" && mario.getY() >= yOben - 45 ) {
        mario.switchBildaktiv("ArrowDown");
    }

    zaehlerUnten1 = zaehlerUnten1 + vZaehler;
    zaehlerUnten2 = zaehlerUnten2 + vZaehler;
    i++;
    if ( i == 50 ) {
        mario.switchBildpassiv();
        i = 0;
    }
    
    //mario.hindernisKollision(hindernis);
    if ( levelMap.kollidiert(mario.getY() ) != 600 ) {
        mario.setLeben(-1);
        //mario.reset(levelMap.kollidiert(mario.getY()));
    }

}

view.setStatusMsg("Ende!");
```
