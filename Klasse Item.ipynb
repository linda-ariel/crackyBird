```c++
//#ifndef ITEM
//#define ITEM

#include <algoviz/SVG.hpp>
#include <iostream>
using namespace std;

class Item {
    
    private:
        int xpos;
        int ypos;
        int art; //3 verschiedene
        bool kollision;
        SVG *spielfeld;
        Image bild;
    
    public:  
        //Standartonstruktor
        Item(int xpos, int ypos, int art, SVG *spielfeld) {
            this->xpos = xpos;
            this->ypos = ypos;
            this->art = art;
            
            if (art == 0) {
                //stein
                bild = Image("", xpos, ypos, 50, 50, spielfeld);
            } else if (art == 1) {
                //münze
                bild = Image("", xpos, ypos, 30, 30, spielfeld);
            } else if (art == 2) {
                //feind
                bild = Image("",xpos, ypos, 60, 100, spielfeld);
            }
        }

    
        //getter
        int getXpos() {
            return xpos;
        }
    
        int getYpos() {
            return ypos;
        }
    
        int getArt() {
            return art;
        }
    
        Image getBild() {
            return bild;
        }
    
    
        //setter
        void setXpos(int xpos) {
            this->xpos = xpos;
        }
    
        void setYpos(int ypos) {
            this->ypos = ypos;
        }
        
        void setArt(int art)  {
            if (art < 0 || art > 2) {
                cout << "Achtung! Dieses Item existiert nicht";
            } else {
                this->art = art;
            }
            
        }        
        
        void setBild(Image bild)  {
            this->bild = bild;
        }
    
    
        //operationen
    
        //Münze
        void münzeGesammelt(Image &münze) { 
                //Münze wird immer kleiner
                for(int size = 30; size > 5; size-- ) {
                    münze.setSize(size,size);
                    AlgoViz::sleep(10);
                }
                münze.removeFromView();
        }
    
        //Stein      
        void steinZerstört(Image &stein) { 
            //Stein zerfällt in vier steine
            Image lo = Image("", stein.getX() - 25, stein.getY() - 25, 25, 25, spielfeld);
            Image ro = Image("", stein.getX() + 25, stein.getY() - 25, 25, 25, spielfeld);
            Image lu = Image("", stein.getX() - 25, stein.getY() + 25, 25, 25, spielfeld);
            Image ru = Image("", stein.getX() + 25, stein.getY() + 25, 25, 25, spielfeld);
            //eigentlicher Stein wird entfernt 
            stein.removeFromView();
            //Steine springen nach außen weg
            for (int zaehler = 0; zaehler < 50; zaehler++) {
                lo.moveTo(lo.getX() - 1, lo.getY() - 1);
                ro.moveTo(ro.getX() + 1, ro.getY() - 1);
                lu.moveTo(lu.getX() - 1, lu.getY() + 1);
                ru.moveTo(ru.getX() + 1, ru.getY() + 1);
                AlgoViz::sleep(5);

            }
                lo.removeFromView();
                ro.removeFromView();
                lu.removeFromView();
                ru.removeFromView();   
        }
            
    
        //Feind
        void feindTot(Image &feind) {
            //Feind wird plattgedrückt
            for (int zaehler = 0; zaehler < 50; zaehler++) {
                feind.setSize (60 + zaehler, 100 - zaehler);
                feind.moveTo(feind.getX(), feind.getY() + 1);
            }
        }
            
            
    
};
//#endif
```
