```c++
#include <algoviz/SVG.hpp>
#include "Level2.hpp"
#include "Protagonist.hpp"
#include "Hindernis.hpp"
#include "Item.hpp"
#include "Itemverwaltung.hpp"

AlgoViz::clear();

SVG view = SVG(700, 500, "Level 1");
Image backround = Image("erste_uni.jpg",350,250,700,500,&view);
view.drawRect(0,0,700,550);
view.drawRect(110,0,80,550);

view.setStatusMsg("Drücken Sie x zu beenden!");

Text anzeigeL = Text("Leben: 3", 15, 20, &view);
Text anzeigeP = Text("Punkte: 0", 15, 37, &view);
view.drawRect(110,380,80,80);


MouseState maus;

//Spieler
Protagonist mario = Protagonist(&view, &anzeigeP, &anzeigeL,1);

Itemverwaltung bib = Itemverwaltung(&view);
bib.addItem(3,0);


//Karte des Levels
Level levelMap = Level(1,5,5,&view);

bool blockade = false; // Einmal gedrückt, ist der Sprung vorerst nicht beeinflussbar...man kann aber "Supersprung" aktivieren
bool blockade2 = false; // Blockiert jegliche Aktionen während des Runterspringens
int arrowUp = 0; // Speichert die Anzahl der "arrowUps", die während der Blockade gedrückt werden...
bool supersprungB = false;
string taste = "";


int i = 0;
int j = 0;
int zaehler = 0;
int yOben;
int yObenFix;
double vZaehler = 1;
double zaehlerUnten1 = 0;
double zaehlerUnten2 = 0;
bool zweihundert = false;

while ( taste != "x" ) {
    
    bib.moveItems(0.5);
    bib.kollision(mario.getY());
    
   /*if ( zaehlerUnten1 >= 400 ) {
        
        if (levelMap.getEbene2() == 1) {
            levelMap.add(1);
        }
        zaehlerUnten1 = 0;
        zweihundert = true;
    }
    
    if ( zaehlerUnten1 >= 200 && zweihundert == true ) {
        
        if (levelMap.getEbene2() == 1) {
            levelMap.add(1);
        }
        zaehlerUnten1 = 0;
        zweihundert = false;
    }*/

    //Wenn die Karte zu ende es, wird sie von vorne wiederbefüllt
    if ( zaehlerUnten2 >= 300 ) {
        
        zaehler++;
        if ( zaehler == 30 ) {
            cout << "Level up!" << endl;
            zaehler = 0;
            levelMap.fuellMap();
            double neuV = levelMap.getV() + 0.25;
            vZaehler = neuV;
            levelMap.setV(neuV);
            mario.setV(neuV);
        }
        if (levelMap.getEbene1() == 1) {
            levelMap.add2(0);
        }
        
        if (levelMap.getEbene2() == 1) {
            levelMap.add2(1);
        }
            
        if (levelMap.getEbene3() == 1) {
            levelMap.add2(2);
        }
        
        zaehlerUnten2 = 0;
 
    }
    
    
    /*für die drei Unterschiedliche Höhe
    for(int j = 0; j < 3; j++){
        
        
        switch(levelMap.getValue(i,j)){
            case 0 : 
                levelMap.move();
                break;
            case 1:
                levelMap.add(j);  
                levelMap.move();
                break;
                
        }*/
        
       /* if (levelMap.getValue(i,j) == 1) { 
            
            levelMap.add(j);  
            levelMap.move();
            
            
        } else {
            
            levelMap.move();
          
        }
    }*/
    
    levelMap.move();
    
    taste = view.lastKey();
    yOben = levelMap.woIstBoden(mario.getY());
    cout << yOben << endl;
    mario.setYoben(yOben);

    if ( taste == "ArrowUp" && blockade2 == false ) {
        if ( blockade == false ) {
            yObenFix = yOben;
            blockade = true;
        }
        
        if ( mario.getBildwert() == 2 ) {
            mario.switchBildaktiv("ArrowUp");
            
        } else { 
           
            mario.springenhoch();
            
            if ( mario.getY() + 70 < yObenFix - 45) { 
                blockade = false;
                blockade2 = true;
            } 
        }
    } else if ( taste == "Shift" && blockade2 == false ) {
        if ( blockade == false ) {
            yObenFix = yOben;
            blockade = true;
            mario.supersprung();
            supersprungB = true;
        }
        
        mario.springenhoch();
        
        if ( mario.getY() + 280 < yObenFix - 45 ) {
            blockade = false;
            blockade2 = true;
        }
            
    } else if ( taste != "ArrowUp" && taste != "Shift" && blockade == true ) {
        
        if ( supersprungB == true ) {
            
            mario.springenhoch();
            
            if ( mario.getY() + 280 < yObenFix - 45 ) {
                blockade = false;
                blockade2 = true;
            }
            
        } else {
            
            mario.springenhoch();
            
            if ( mario.getY() + 70 < yObenFix - 45 ) {
                blockade = false;
                blockade2 = true;
            }
        }
    
    } else if ( (blockade2 == true) || ( blockade == false && blockade2 == false ) ) {
     
        if ( mario.getY() < yOben - 45 ) {
            mario.springenrunter();
        }
    
        if ( mario.getY() >= yOben - 45 ) {
            blockade2 = false;
            supersprungB = false;
            mario.setSupersprung(false);
        }
    
    } else if ( taste == "ArrowDown" && blockade == false && blockade2 == false ) {
        mario.switchBildaktiv("ArrowDown");
    }

    zaehlerUnten1 = zaehlerUnten1 + vZaehler;
    zaehlerUnten2 = zaehlerUnten2 + vZaehler;
    i++;
    if ( i == 50 ) {
        mario.switchBildpassiv();
        i = 0;
    }
    
    //mario.hindernisKollision(hindernis);
    if ( levelMap.kollidiert(mario.getY() ) != 600 ) {
        mario.setLeben(-1);
        //mario.reset(levelMap.kollidiert(mario.getY()));
    }

}

view.setStatusMsg("Ende!");
```
