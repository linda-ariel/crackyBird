```c++
#include <algoviz/SVG.hpp>
#include "Level.hpp"
#include "Protagonist.hpp"
#include "Item.hpp"
#include "Itemverwaltung.hpp"

using namespace std;

MouseState maus;

AlgoViz::clear();

// Regel werden erklärt: Folie 1
SVG regeln1 = SVG(700, 500);

Rect hintergrund = Rect(297,247,6,6,&regeln1);
int xKoord = 347;
int yKoord = 247;
int weite = 6;
hintergrund.setColor("transparent");
hintergrund.setFill(48,92,232,0.1);
double transparenz = 0.1;
for ( int i = 0; i < 70; i++ ){
    weite = weite + 10;
    xKoord = xKoord - 5;
    yKoord = yKoord - 5;
    hintergrund.setWidth(weite);
    hintergrund.setHeight(weite);
    hintergrund.moveTo(xKoord,yKoord);
    transparenz = transparenz + 0.01;
    hintergrund.setFill(48,92,232,transparenz);
    AlgoViz::sleep(5);
}
Rect button = Rect(200,440,300,40,3,3,&regeln1);
button.setFill(45,243,196,0.9);

regeln1.drawText("So wird", 150, 42, 30, "century");
regeln1.drawText("FlappyBird", 285, 23, 20, "century");
regeln1.drawText("ohne Bird", 300, 41, 15, "century");
regeln1.drawText("mit Student", 282, 60, 20, "century");
regeln1.drawText("gespielt...", 420, 42, 30, "century");
regeln1.drawText("1) Hindernisse sind bunte Blöcke, die im gesamten", 50, 100, 20, "century");
regeln1.drawText("Spielfeld erscheinen können: ", 70, 140, 20, "century");
regeln1.drawText("...Umgehe die Hindernisse, indem du auf bzw. über sie", 70, 180, 20, "century");
regeln1.drawText("springst oder auch unter ihnen hindurchkriechst", 70, 220, 20, "century");
regeln1.drawText("2) Du möchtst ECTS sammeln? (JAAAAA!)...", 50, 260, 20, "century");
regeln1.drawText("...Sammle so viele Münzen wie möglich ein", 70, 300, 20, "century");
regeln1.drawText("...Zerstöre die Steine, die dir in den Weg gelegt", 70, 340, 20, "century");
regeln1.drawText("werden und verdiene dir gleich 3 Credits auf einmal! :D", 70, 380, 20, "century");
regeln1.drawText("...Stolperst du aber über die Steine, verlierst du 3 Credits :(", 70, 420, 20, "century");
regeln1.drawText("So weit, so gut", 245, 470, 30, "century");

// Zur Bestätigung muss der hellblaue Knopf gedrückt werden
do {
    maus = regeln1.waitForClick();
} while ( maus.isLegal() == false || maus.x() < 200 || maus.x() > 500 || maus.y() < 440 || maus.y() > 480 );

AlgoViz::clear();

// Regel werden erklärt: Folie 2
SVG regeln2 = SVG(700, 500);

Rect hintergrund1 = Rect(0,0,700,500,&regeln2);
hintergrund1.setColor("transparent");
hintergrund1.setFill(48,92,232,0.7);
Rect button1 = Rect(200,440,300,40,3,3,&regeln2);
button1.setFill(45,243,196,0.9);

regeln2.drawText("So wird", 150, 52, 30, "century");
regeln2.drawText("FlappyBird", 285, 33, 20, "century");
regeln2.drawText("ohne Bird", 300, 51, 15, "century");
regeln2.drawText("mit Student", 282, 70, 20, "century");
regeln2.drawText("gespielt...", 420, 52, 30, "century");
regeln2.drawText("4) Pass auf deine schlimmsten Feinde auf, PRÜFUNGEN(!):", 50, 140, 20, "century");
regeln2.drawText("Sie sind fett, aggressiv und wollen dich umbringen...", 70, 180, 20, "century");
regeln2.drawText("...Also töte sie zuerst und du erhälst ein Leben!", 70, 220, 20, "century");
regeln2.drawText("...Andererseits verlierst du eins", 70, 260, 20, "century");
regeln2.drawText("5) Überlebst du ein Semester, dann kommst du ins nächste!", 50, 300, 20, "century");
regeln2.drawText("...Es wird also alles noch etwas kniffliger", 70, 340, 20, "century");
regeln2.drawText("Hab ich verstanden!", 210, 470, 30, "century");

do {
    maus = regeln2.waitForClick();
} while ( maus.isLegal() == false || maus.x() < 200 || maus.x() > 500 || maus.y() < 440 || maus.y() > 480 );

AlgoViz::clear();

// Steuerung wird erklärt
SVG steuerung = SVG(700, 500);

Rect hintergrund2 = Rect(0,0,700,500,&steuerung);
hintergrund2.setColor("transparent");
hintergrund2.setFill(48,92,232,0.7);
Rect button2 = Rect(200,440,300,40,3,3,&steuerung);
button2.setFill(45,243,196,0.9);


steuerung.drawText("Ein bisschen was zur Steuerung...", 100, 50, 30, "century");
steuerung.drawText("1) Sprung: Drücke die Pfeiltaste nach OBEN, wenn du einen", 30, 100, 20, "century");
steuerung.drawText("kleinen Sprung machen möchtest", 50, 140, 20, "century");
steuerung.drawText("2) Supersprung: Drücke SHIFT, um einen groooßen Sprung zu machen", 30, 180, 20, "century");
steuerung.drawText("3) Kriechen: Drücke die Pfeiltaste nach UNTEN, wenn du unter", 30, 220, 20, "century");
steuerung.drawText("einem Hindernis hindurchkriechen möchtest und drücke dann die", 50, 260, 20, "century");
steuerung.drawText("Pfeiltaste nach OBEN, um wieder mit dem Laufen zu beginnen", 50, 300, 20, "century");
steuerung.drawText("4) Zerstörer-Modus: Drücke LEERTASTE, wenn du einen Feind oder", 30, 340, 20, "century");
steuerung.drawText("einen Stein aus dem Weg räumen willst...", 50, 380, 20, "century");
steuerung.drawText("Aber PASS AUF!: Deine Verwandlung ist nicht von Dauer", 50, 420, 20, "century");
steuerung.drawText("Kann's los gehen?!", 220, 470, 30, "century");

do {
    maus = steuerung.waitForClick();
} while ( maus.isLegal() == false || maus.x() < 200 || maus.x() > 500 || maus.y() < 440 || maus.y() > 480 );

AlgoViz::clear();

/*
 *
 * Beginn des "eigentlichen" Spiels
 *
 */

// SVG und Hintergrund werden erstellt
SVG view = SVG(700, 500, "Level 1");
Image backround = Image("PizzaBoxHell.jpg",350,250,700,500,&view);

// Anzeigen für Leben sowie für gesammelte ECTS
Text anzeigeL = Text("Leben: 3", 15, 20, &view);
Text anzeigeP = Text("ECTS: 0", 15, 37, &view);

// Kurze Benachrichtigung, wenn die nächste Geschwindigkeitsstufe erreicht wird (nächstes Level)
Text levelUp = Text("!!Achtung!! Es wird schneller", 290, 100, &view);
levelUp.setColor("red");
levelUp.hide();

// "Zerstörer-Modus": Balken -> Anzeige der Dauer und Regenerierungsphase
Rect coolDown = Rect (450,5,200,28, &view);
coolDown.setFill("LightBlue");
int coolDownTime = 0;

// Parameter für den Level-Konstruktor
int geschwindigkeit;
int muenze = 10;
int feind = 5;
int stein = 3;

// Festlegung der Anfangsgeschwindigkeit durch den Nutzer
cout << "Gebe die Geschwindigkeit (zwischen 1-20) an, mit der du in das Spiel starten möchtest" << endl;
cout << "(!Schnelligkeit kann je nach Computer-Leistung sehr stark variieren!)" << endl;
do {
    cin >> geschwindigkeit;
} while ( geschwindigkeit < 1 && geschwindigkeit > 20 );

//Karte des Levels
Level levelMap = Level(geschwindigkeit,muenze,feind,stein,&view);

//Spieler
Protagonist mario = Protagonist(&view, &anzeigeP, &anzeigeL, geschwindigkeit);

//Items
Itemverwaltung bib = Itemverwaltung(geschwindigkeit,&view);


// Blockade gibt an, ob der Spieler gerade hochspringt
bool blockade = false; 

// Blockade gibt an, ob sich der Spieler gerade (nach einem Sprung) im Fall befindet
bool blockade2 = false;

// (In-)Aktivierung des höheren Sprungs
bool supersprungB = false;


// Zähler...

// ...für den "passiven" Bildwechsel
int i = 0;

// ...für die Abarbeitung der Blöcke pro Level
int zaehler = 0;

// ...für die "Unverwundbarkeit" nach einer Kollision mit einem Hindernis
int kollisionZ = 0;

// ...für die "Unverwundbarkeit" nach einer Kollision mit einem Stein/Feind
int safety = 0;

// ..., der die aktuelle Geschwindigkeit angibt
int vZaehler = 1;

// ...für die Einhaltung gewisser Abstände beim Setzen von Hindernissen/Items
// in Abhängigkeit der Geschwindigkeit
int abstandZ = 0;

// ...für die Zeit, in der der Spieler im Zerstörer-Modus ist
int keinBildwechselZ = 0;



// nächstes Hindernis bzw. Boden unter dem Spieler (y-Koordinate)
int yOben;

// festgehaltene Hindernis- bzw. Bodenhöhe beim Absprung
int yObenFix;


string taste = "";


cout << "Klicke auf das Spielfeld, um zu starten" << endl;
maus = view.waitForClick();


//die Hauptschleife
while ( mario.getLeben() > 0 && taste != "x") {
    
    // mögliche neue Hindernisse/Items können auf das Spielfeld gesetzt werden
    if ( abstandZ >= 300 ) {
        
        levelUp.hide();
        zaehler++;
        
        // nächstes Level wird erreicht und somit konfiguriert
        if ( zaehler == 30 ) {
            
            cout << "Level up!" << endl;
            
            zaehler = 0;
            
            muenze++;
            feind++;
            stein++;
            levelMap.buildLevel(muenze, feind, stein);
            
            levelUp.show();
            
            int neuV = levelMap.getV() + 1;
            vZaehler = neuV;
            levelMap.setV(neuV);
            mario.setV(neuV);
            bib.setV(neuV);
            
        }
        
       //Items und Hindernisse hinzufügen   
       switch (levelMap.getEbene1()){
               
           case 0 :
               break;
        
           case 1:  
               
               levelMap.add(0);
               break;
               
           case 2: 
               
               bib.addItem(0,1);
               break;
        
           case 3:
               
               bib.addItem(0,2);
               break;
               
        
           case 4:
               
               bib.addItem(0,0);
               break;
       }
        
       switch (levelMap.getEbene2()){
               
           case 0 :
               break;
                          case 1:  
               
               levelMap.add(1);
               break;
               
           case 2: 
               
               bib.addItem(1,1);
               break;
        
           case 3:
               
               bib.addItem(1,2);
               break;
               
        
           case 4:
               
               bib.addItem(1,0);
               break;
         
       }
        
       switch (levelMap.getEbene3()){
               
           case 0 :
                   break;
        
           case 1:  
               
               levelMap.add(2);
               break;
               
           case 2: 
               
               bib.addItem(2,1);
               break;
        
           case 3:
               
               bib.addItem(2,2);
               break;
               
        
           case 4:
               
               bib.addItem(2,0);
               break; 
        }
        
            abstandZ = 0;
    }
    
    //Items/Hindernisse bewegen
    levelMap.move();
    bib.moveItems();

    taste = view.lastKey();
    yOben = levelMap.woIstBoden(mario.getY());
    mario.setYoben(yOben);
    
    
    //Steuerung des Protagonisten
    
    // Beginn eines kleinen Sprunges oder Aufstellen nach "Kriechen"
    if ( taste == "ArrowUp" && blockade2 == false ) {
        
        if ( mario.getBildwert() == 2 ) {
            mario.switchBildaktiv("ArrowUp");
            
        } else {
            
            if ( blockade == false ) {
            yObenFix = yOben;
            blockade = true;
            }
            
            mario.springenhoch();
            
            if ( mario.getY() + 70 < yObenFix - 50) { 
                blockade = false;
                blockade2 = true;
            }
        }
        
    // großer Sprung
    } else if ( taste == "Shift" && blockade2 == false && mario.getBildwert() != 2 ) {
        
        if ( blockade == false ) {
            yObenFix = yOben;
            blockade = true;
            mario.supersprung();
            supersprungB = true;
        }
        
        mario.springenhoch();
        
        if ( mario.getY() + 280 < yObenFix - 50 ) {
            blockade = false;
            blockade2 = true;
        }
    
    // Fortsetzung des eingeleiteten Sprungvorgangs
    } else if ( taste != "ArrowUp" && taste != "Shift" && blockade == true ) {
        
        if ( taste == " " && coolDownTime == 0 ){
            if ( mario.getBildwert() != 3 ) {
                mario.switchBildaktiv("fight");
            }
        }
        
        if ( supersprungB == true ) {
            
            mario.springenhoch();
            
            if ( mario.getY() + 280 < yObenFix - 50 ) {
                blockade = false;
                blockade2 = true;
            }
            
        } else {
            
            mario.springenhoch();
            
            if ( mario.getY() + 70 < yObenFix - 50 ) {
                blockade = false;
                blockade2 = true;
            }
        }
    
    // Runterspringen/Fall des Spielers
    } else if ( (blockade2 == true) || ( blockade == false && blockade2 == false && taste != "ArrowDown") ) {
     
        if ( taste == " " && coolDownTime == 0 ) {
            if ( mario.getBildwert() != 3 ) {
                mario.switchBildaktiv("fight");
            }
        }
        
        if ( mario.getY() < yOben - 50 ) {
            mario.springenrunter();
            //cout << mario.getY() << endl;
        }
    
        if ( mario.getY() >= yOben - 50 ) {
            blockade2 = false;
            supersprungB = false;
            mario.setSupersprung(false);
        }
    
    // Beginn des Kriechvorgangs
    } else if ( taste == "ArrowDown" && blockade == false && blockade2 == false ) {
        
        mario.switchBildaktiv("ArrowDown");
        
    }

    // Zähleraktualisierung
    abstandZ = abstandZ + vZaehler;
    i++;
    
    // "passiver" Bildwechsel (beim Laufen oder Aufkommen auf dem Boden)
    if ( i == 50 ) {
        i = 0;
        mario.switchBildpassiv();
    }
    
    //Zerstörer-Modus: Balken-Aktualisierung
    if ( mario.getBildwert() == 3 ) {
        
        keinBildwechselZ++;
        coolDown.setWidth(coolDown.getWidth() - 1);
        
        if ( keinBildwechselZ == 200 ) {
            coolDownTime = 200;
            coolDown.setFill("red");
            keinBildwechselZ = 0;
            mario.switchBildaktiv("OutOfFight");
        }
    }
    
    if ( coolDownTime > 0 ) {
        coolDownTime--;
        coolDown.setWidth(coolDown.getWidth() + 1);
        if ( coolDownTime == 0 ) {
            coolDown.setFill("LightBlue");
        }
    }
    
    // Hinderniskollision
    if ( kollisionZ > 0 ) {
        kollisionZ++;
        if ( kollisionZ == 250 ) {
            kollisionZ = 0;
            if ( safety == 0 ) {
                mario.setBlinken(false);
            }
        }
    }

    if ( kollisionZ == 0 ) {
        int rueckgabe = levelMap.kollidiert(mario.getY(), mario.getBildwert());
        
        if ( rueckgabe != 600 ) {
            if ( mario.getY() < rueckgabe || rueckgabe >= 440 ) {
                mario.moveTo(rueckgabe - 50);
            } else {
                blockade = false;
                blockade2 = true;
            }
            mario.setLeben(-1);
            mario.setBlinken(true);
            kollisionZ++;
        }
    }
    
    
    //Kollision mit Items 
      
    if ( safety == 0 ) {
        
        int itemkollision = bib.kollision(mario.getY(), mario.getBildwert());

            switch (itemkollision){

               case 0:

                    break;

               case -3:  

                   mario.setPunkte(-3);
                   safety++;
                   break;

               case -1: 

                   mario.setLeben(-1);
                   mario.setBlinken(true);
                   safety++;
                   break;

               case 1:

                   mario.setLeben(1);
                   safety++;
                   break;


               case 2:

                   mario.setPunkte(1);
                   break; 

                case 3: 

                   mario.setPunkte(3);
                   safety++;
                   break;
            }
    
    } else {
        
        safety++;
        
        if ( safety == 130 ) {
            
            safety = 0;
            
            if ( kollisionZ == 0 ) {
                mario.setBlinken(false);
            }
        }
    }
    
}


// Game Over
view.setStatusMsg("Game Over!");
Rect schwarz = Rect (0,0,700,550,&view);
schwarz.setFill(48,92,232);
Image tot = Image("gameOver2.png",340,400,200,200,&view);
view.drawText("GAME OVER", 210, 150, 40, "century");
view.drawText("ECTS: " + to_string(mario.getPunkte()), 285, 230, 25, "century");
if (mario.getPunkte() < 10 ) {
    view.drawText("Du solltest dich lieber umorientieren...", 170, 270, 21, "century");
    view.drawText("Vielleicht liegt dir ein anderes Studium ja besser", 115, 300, 21, "century");
} else if (mario.getPunkte() >= 10 && mario.getPunkte() < 30) {
    view.drawText("Du musst wohl noch etwas an deiner", 160, 270, 21, "century");
    view.drawText("Arbeitsmoral ändern", 235, 300, 21, "century");
} else if (mario.getPunkte() >= 30 && mario.getPunkte() < 50) {
    view.drawText("Weiter so, du kannst das packen!", 180, 270, 21, "century");
    view.drawText("...im nächsten Anlauf", 230, 300, 21, "century");
} else if (mario.getPunkte() >= 50) {
    view.drawText("Wirklich gut!", 275, 270, 21, "century");
    view.drawText("Du hast dich sehr wacker gehalten", 170, 300, 21, "century");
}
```
