```c++
#ifndef _HINDERNIS_
#define _HINDERNIS_

#include <algoviz/SVG.hpp>
#include <cstdlib>

using namespace std;

class Hindernisstrecke {
    
 private:
  vector<*Rect> strecke;
  Rect* block;

  // geschwindigkeit
  int v;

 public:
  // konstruktor
  Hindernisstrecke() {}

  // getter
  void getV() { return v; }

  // setter
  void setV(int v) { this->v = v; }

  // bewegt Rect nach links
  void move() {
    int neuX;
    auto it = strecke.begin();

    // f√ºr jede leere gehen wir 50 pixels
    for (int j = 0; j < 5; j++) {
      while (it != strecke.end()) {
        // hole ein zeiger
        Rect* block = *it;

        // hole seine position
        int x = block->getX();
        int y = block->getY();

        // berechne neue position
        neuX = x - v;

        // bewege block
        block->moveTo(neuX, y);
        AlgoViz::sleep(5);

        it++;

        // Falls block nicht mehr zusehen ist
        // if(neuX < 0){
        //  hindernisstrecke.erase(it);
        // delete block;
        //}
      }
    }
  }
    
    void add(Rect r){
        strecke.push_back(r);
    }
    
    
};
#endif
```
