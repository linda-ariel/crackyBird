```c++
#include <algoviz/SVG.hpp>
#include "Level.hpp"
#include "Protagonist.hpp"
#include "Hindernis.hpp"

AlgoViz::clear();

SVG view = SVG(700, 500, "Level 1");
Image backround = Image("erste_uni.jpg",350,250,700,500,&view);
view.drawRect(0,0,700,550);
view.drawRect(110,0,80,550);

view.setStatusMsg("Drücken Sie x zu beenden!");

Text anzeigeL = Text("Leben: 3", 15, 20, &view);
Text anzeigeP = Text("Punkte: 0", 15, 37, &view);
view.drawRect(110,380,80,80);


MouseState maus;

//Spieler
Protagonist mario = Protagonist(&view, &anzeigeP, &anzeigeL);


//Karte des Levels
Level levelMap = Level(2,5,5,5,&view);

bool blockade = false; // Einmal gedrückt, ist der Sprung vorerst nicht beeinflussbar...man kann aber "Supersprung" aktivieren
bool blockade2 = false; // Blockiert jegliche Aktionen während des Runterspringens
int arrowUp = 0; // Speichert die Anzahl der "arrowUps", die während der Blockade gedrückt werden...
bool supersprungB = false;
string taste = "";


int i = 0;
int zaehler = 0;
int blockZaehler = 0;
int yOben;
list<Rect*> hindernis;
bool geloescht = false;

levelMap.print();
while ( mario.getLeben() > 0 && taste != "x" ) {
    

    //Wenn die Karte zu ende es, wird sie von vorne wiederbefüllt
    if (i == 100) {
        
        for (int j = 0; j < 50; j++) {
            
            levelMap.move();
            
            }
        
            levelMap.fuellMap();
            double neuV = levelMap.getV() + 0.25;
            levelMap.setV(neuV);
            i = 0;
 
    }
    
    
    //für die drei Unterschiedliche Höhe
    for(int j = 0; j < 3; j++){
        
        
        switch(levelMap.getValue(i,j)){
            case 0 : 
                levelMap.move();
                break;
            case 1:
                levelMap.add(j);  
                levelMap.move();
                break;
                
        }
        
       /* if (levelMap.getValue(i,j) == 1) { 
            
            levelMap.add(j);  
            levelMap.move();
            
            
        } else {
            
            levelMap.move();
          
        }*/
    }
    
    levelMap.move();
    
    taste = view.lastKey();
    yOben = levelMap.woIstBoden(mario.getY());

    if ( taste == "ArrowUp" && blockade == false ) {
        if ( mario.getBildwert() == 2 ) {
            mario.switchBildaktiv("ArrowUp");
        } else if ( supersprungB == false && mario.getY() + 140 >= yOben - 40 ) { 
            mario.springenhoch();
            blockade = true; 
        }
    } else if ( taste == "ArrowUp" && blockade == true ) {
        if ( mario.getY() + 140 >= yOben - 40 && supersprungB == false && blockade2 == false ) {
            mario.springenhoch();
            arrowUp++;
            if ( mario.getY() + 140 < yOben - 40 && arrowUp >= 4 ) { 
                arrowUp = 0;
                supersprungB = true;
                mario.supersprung();
            } else if ( mario.getY() + 140 < yOben - 40 && arrowUp < 4 ) {
                arrowUp = 0;
                blockade2 = true;
            } 
        } else if ( mario.getY() + 280 >= yOben - 40 && supersprungB == true && blockade2 == false ) {
            mario.springenhoch();
            if ( mario.getY() + 280 < yOben - 40 ) {
                blockade2 = true;
            }
        } else if ( mario.getY() < yOben - 40 && blockade2 == true ) {
            mario.springenrunter();
            if ( mario.getY() >= yOben - 40 ) {
                blockade2 = false;
                blockade = false;
                supersprungB = false;
                mario.setSupersprung(false);
            }
        }
    } else if ( taste != "ArrowUp" && blockade == true ) {
        if ( mario.getY() + 140 >= yOben - 40 && supersprungB == false && blockade2 == false ) {
            mario.springenhoch();
            if ( mario.getY() + 140 < yOben - 40 && arrowUp >= 4 ) {
                arrowUp = 0;
                supersprungB = true;
                mario.supersprung();
            } else if ( mario.getY() + 140 < yOben - 40 && arrowUp < 4 ) {
                arrowUp = 0;
                blockade2 = true;
            }
        } else if ( mario.getY() + 280 >= yOben - 40 && supersprungB == true && blockade2 == false ) {
            mario.springenhoch();
            if ( mario.getY() + 280 < yOben - 40 ) {
                blockade2 = true;
            }
        } else if ( mario.getY() < yOben - 40 && blockade2 == true ) {
            mario.springenrunter();
            if ( mario.getY() >= yOben - 40 ) {
                blockade2 = false;
                blockade = false;
                supersprungB = false;
                mario.setSupersprung(false);
            }
        }
    } else if ( taste == "ArrowDown" && mario.getY() >= yOben - 40 ) {
        mario.switchBildaktiv("ArrowDown");
    }

    i++;
    if ( i % 10 == 0 ) {
        mario.switchBildpassiv();
    }
    
    //mario.hindernisKollision(hindernis);
    if ( levelMap.kollidiert(mario.getY() )) {
        //mario.setLeben(-1);
    }

}

view.setStatusMsg("Ende!");
```
